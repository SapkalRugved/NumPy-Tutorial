# 1. Eigen Vector & Eigen Values

```python
# Define a 2x2 matrix
A = np.array([[5,4],
              [1,2]])

# np.linalg.eig() computes the eigenvalues and eigenvectors of matrix A
# It returns two things:
# 1. eigenvalues  -> the λ values
# 2. eigenvectors -> the vectors corresponding to each λ
eigenvalues, eigenvectors = np.linalg.eig(A)

# Print eigenvalues and eigenvectors
eigenvalues, eigenvectors
```
