```python
import numpy as np
```

# 1. Generating Random Numbers

```python
print(np.random.rand())
print(np.random.rand(3))

print(np.random.randint(10))
print(np.random.randint(10, size=(3, 5)))
```

# 2. Random Numbers with Seed

```python
# np.random.seed(seed): Sets global random seed for REPRODUCIBLE results - SAME sequence every run
# Parameters: integer seed value (e.g., 2) - **CRITICAL for ML debugging and testing**
# Speed: O(1) operation - just resets internal state, no number generation
np.random.seed(2)

# ALWAYS same output after seed(2): 0.37454011884736244 ∈ [0, 1)
print(np.random.rand())             

# ALWAYS same 1D array after seed(2): [0.95071431 0.73199394 0.59865848]
print(np.random.rand(3))            

# ALWAYS same int after previous calls: 7 ∈ [0, 10)
print(np.random.randint(10))        

# ALWAYS same 3×5 matrix after previous calls - perfect for reproducible testing
print(np.random.randint(10, size=(3, 5)))
```

```python
np.random.seed(5)

print(np.random.rand())
print(np.random.rand(3))

print(np.random.randint(10))
print(np.random.randint(10, size=(3, 5)))
```

# 3. Seed in Loops

```python
np.random.seed(5)  # Set seed for reproducible results - same sequence every run

for i in range(10):
  arr = np.random.randint(2, size=1000)  # Generate 1000 binary ints (0 or 1) ∈ [0, 2)
  print(len(arr[arr == 1]) , len(arr[arr == 0]))  # Count 1s and 0s in array
```
