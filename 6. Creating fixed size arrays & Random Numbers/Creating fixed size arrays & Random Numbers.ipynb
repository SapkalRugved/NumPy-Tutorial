```python
import numpy as np
```

# 1. Zeros

```python
#Create a 1D array of 10 zeros (shape: (10,))
zeros_array = np.zeros(10)
print(zeros_array)
```

```python
# Create a 2D array (matrix) of 2 rows and 4 columns, all zeros (shape: (2, 4))
zeros_array = np.zeros((2, 4))
print(zeros_array)
```

```python
# Create a 3D array with shape (2, 3, 4) - think of it as 2 matrices of 3x4
zeros_array = np.zeros((2, 3, 4))
print(zeros_array)
```

```python
# Create a 2D integer array (2x3) instead of default float64
zeros_array = np.zeros((2, 3), dtype=int)
print(zeros_array)
```

# 2. One

```python
# Create a 1D array of 10 ones (shape: (10,))
# Default dtype is float64
ones_array = np.ones(10)
print(ones_array)
```

```python
# Create a 2D array (matrix) of 2 rows and 4 columns, all ones (shape: (2, 4))
ones_array = np.ones((2, 4))
print(ones_array)
```

```python
# Create a 3D array with shape (2, 3, 4) - 2 blocks of 3x4 matrices of ones
ones_array = np.ones((2, 3, 4))
print(ones_array)
```

```python
# Create a 2D integer array (2x3) filled with ones (explicit int dtype)
ones_array = np.ones((2, 3), dtype=int)
print(ones_array)
```

# 3. Eye

```python
# Create 3x3 Identity matrix (square diagonal matrix)
# Diagonal = 1s, off-diagonal = 0s (shape: (3, 3))
eye_array = np.eye(3)
print(eye_array)
```

```python
# Create 3x4 Identity-like matrix (rectangular)
# 3 rows, 4 columns - diagonal 1s up to min(3,4)=3, rest 0s (shape: (3, 4))
eye_array = np.eye(3, 4)
print(eye_array)
```

```python
# Create 3x4 Identity matrix with integer dtype instead of float64
eye_array = np.eye(3, 4, dtype=int)
print(eye_array)
```

# 4. Random Number Generation

```python
# np.random.rand(d0, d1, ..., dn): Returns array of random floats from uniform distribution [0, 1)
# Parameters: 
# - Shape arguments (like np.zeros): ()=scalar, (N)=1D array, (M,N)=2D matrix, etc.
# - Range is ALWAYS fixed [0, 1) - no low/high parameters needed
print(np.random.rand())           # Scalar: single random float ∈ [0, 1)
print(np.random.rand(10))         # 1D array: 10 random floats ∈ [0, 1)
print(np.random.rand(3, 3))       # 2D array: 3×3 matrix of random floats ∈ [0, 1)
```

```python
# np.random.randint(low, high, size, dtype): Returns random integers from [low, high) range
# Parameters: 
# - low: inclusive lower bound (default=0)
# - high: exclusive upper bound 
# - size: output shape (scalar if omitted)
# - dtype: data type (default=int64)
print(np.random.randint(100))                    # Scalar: single int ∈ [0, 100)
print(np.random.randint(100, size=10))           # 1D array: 10 ints ∈ [0, 100)
print(np.random.randint(100, size=(3, 3)))       # 2D array: 3×3 matrix ∈ [0, 100)
print(np.random.randint(low=100, high=200, size=(3, 3)))  # 2D array: 3×3 matrix ∈ [100, 200)
```

```python
# np.random.randn(d0, d1, ..., dn): Returns random floats from standard normal distribution (Gaussian)
# Parameters: 
# - Shape arguments (like np.zeros): ()=scalar, (N)=1D array, (M,N)=2D matrix, etc.
# - Mean = 0, Standard Deviation = 1 (standard normal), bell-shaped curve centered at 0
print(np.random.randn())         # Scalar: single random float ∼ N(0,1) (can be negative/positive)
```

```python
# np.random.uniform(low, high, size): Returns random floats from uniform distribution [low, high)
# Parameters: 
# - low: inclusive lower bound (default=0.0)
# - high: exclusive upper bound (default=1.0) 
# - size: output shape (scalar, array, matrix, etc.)
print(np.random.uniform(size=(3, 3)))    # 2D array: 3×3 matrix ∈ [0, 1)
```

```python
# np.random.binomial(n, p, size): Returns random integers from binomial distribution B(n,p)
# Parameters: 
# - n: number of trials (fixed)
# - p: success probability per trial (0≤p≤1)
# - size: output shape (scalar, array, matrix, etc.)
print(np.random.binomial(10, 0.5, size=(3, 3)))    # 2D array: 3×3 matrix ∼ Binomial(n=10, p=0.5)
```
