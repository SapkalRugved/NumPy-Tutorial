```python
import numpy as np
```

`1. reading from .txt`

```python
data = np.loadtxt('data.txt')
print(data)
```

`2. reading from .npy`

```python
data = np.load('data.npy')
print(data)
```

`3. reading from .csv`

```python
data = np.loadtxt('data.csv', delimiter = ',')
print(data)
```

```python
data = np.loadtxt('data.json', delimiter = ',')
print(data)
```

```python
data = np.loadtxt('data.yaml', delimiter = ',')
print(data)
```
